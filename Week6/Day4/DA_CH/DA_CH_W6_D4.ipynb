```python
import numpy as np
import pandas as pd

# Seed for reproducibility
np.random.seed(0)

# Number of entries
n_entries = 1000

# Simulating dataset
time = np.linspace(0, 100, n_entries)  # Time variable
temperature = 20 + 5 * np.sin(np.pi * time / 50) + np.random.normal(0, 0.5, n_entries)  # Temperature (with some noise)
pressure = 1013 + 20 * np.cos(np.pi * time / 25) + np.random.normal(0, 1, n_entries)  # Pressure (with some noise)
chemical_concentration = 5 + 2 * np.sin(np.pi * time / 10) + np.random.normal(0, 0.2, n_entries)  # Chemical concentration

# Creating DataFrame
ninja_data = pd.DataFrame({
    'Time': time,
    'Temperature': temperature,
    'Pressure': pressure,
    'Chemical Concentration': chemical_concentration
})

# Display the first 5 rows
print(ninja_data.head())
```

```python
cols_to_normalize = ['Temperature', 'Pressure', 'Chemical Concentration']

means = ninja_data[cols_to_normalize].mean()
stds = ninja_data[cols_to_normalize].std()

ninja_data_normalized = ninja_data.copy()
ninja_data_normalized[cols_to_normalize] = (ninja_data[cols_to_normalize] - means) / stds

print(ninja_data_normalized.head())
```

```python
print(ninja_data.describe())
```

```python
corr_matrix = ninja_data.corr()
print(corr_matrix)
```

```python
temp_fft = np.fft.fft(ninja_data['Temperature'])
temp_freq = np.fft.fftfreq(n_entries, d=(time[1] - time[0]))

pressure_fft = np.fft.fft(ninja_data['Pressure'])
pressure_freq = np.fft.fftfreq(n_entries, d=(time[1] - time[0]))

chem_fft = np.fft.fft(ninja_data['Chemical Concentration'])
chem_freq = np.fft.fftfreq(n_entries, d=(time[1] - time[0]))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
sns.pairplot(ninja_data)
plt.show()
```

```python
# Function to plot FFT
def plot_fft(freq, fft_values, title):
    plt.figure(figsize=(10,6))
    # Since FFT is symmetrical, take the positive half
    idx = np.argsort(freq)
    freq = freq[idx]
    fft_values = fft_values[idx]
    positive_freqs = freq[freq >= 0]
    magnitudes = np.abs(fft_values[freq >= 0])

    plt.plot(positive_freqs, magnitudes)
    plt.title(title)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()

# Plotting FFT for Temperature
plot_fft(temp_freq, temp_fft, 'FFT of Temperature')

# Plotting FFT for Pressure
plot_fft(pressure_freq, pressure_fft, 'FFT of Pressure')

# Plotting FFT for Chemical Concentration
plot_fft(chem_freq, chem_fft, 'FFT of Chemical Concentration')
```
