ðŸŒŸ Exercise 1 : Matrix Operations
Instructions
In this exercise, youâ€™ll work with a 3x3 matrix. Hereâ€™s a brief explanation of the concepts:

Determinant: The determinant is a value that can be computed from the elements of a square matrix. It provides important information about the matrix, such as whether it has an inverse, and is used in various areas like linear algebra and calculus. To understand more about it you can watch this video.
Inverse of a Matrix: The inverse of a matrix is a matrix that, when multiplied with the original matrix, results in an identity matrix. Not all matrices have inverses. The inverse is crucial in solving systems of linear equations.
Create a 3x3 matrix and perform the following operations:

Calculate the determinant.
Find the inverse of the matrix.

```python
import numpy as np

A = np.array([[1, 2, 3],
              [0, 1, 4],
              [5, 6, 0]])

determinant = np.linalg.det(A)
print("Determinant of A:", determinant)

inverse = np.linalg.inv(A)
print("Inverse of A:")
print(inverse)
```

Exercise 6 : Basic Hypothesis Testing
Instructions
Create a sample dataset to test the effectiveness of a new training program on employee productivity:


import numpy as np

# Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# Your task is to formulate a hypothesis regarding the training program's effectiveness
# and test it using basic statistical functions in NumPy.


Given a dataset, formulate a simple hypothesis and test it using basic statistical functions in NumPy.



```python
import numpy as np

# Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# Your task is to formulate a hypothesis regarding the training program's effectiveness
# and test it using basic statistical functions in NumPy.

productivity_difference = productivity_after - productivity_before

mean_difference = np.mean(productivity_difference)
print("Mean Difference:", mean_difference)

std_difference = np.std(productivity_difference, ddof=1)
print("Standard Deviation of Differences:", std_difference)

n = len(productivity_difference)
print("Number of Observations:", n)

t_statistic = mean_difference / (std_difference / np.sqrt(n))
print("Calculated t-statistic:", t_statistic)

#The taining program has increased employee productivity.
```

ðŸŒŸ Exercise 8 : Time Series Data Manipulation
Instructions
Generate time series data for the year 2023. Demonstrate slicing for the following intervals:

January to March
April to June
July to September
October to December
Generate a time series data for a specific period and demonstrate how to slice this data for different intervals.

```python
date_range = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')

data = np.random.randint(1, 100, size=len(date_range))

df = pd.DataFrame({'Date': date_range, 'Value': data})

df.set_index('Date', inplace=True)

print("Time Series Data for 2023:")
print(df.head())
```

```python
jan_to_mar = df['2023-01-01':'2023-03-31']

print("\nData from January to March:")
print(jan_to_mar.head())
print("...")
print(jan_to_mar.tail())
```

```python
apr_to_jun = df['2023-04-01':'2023-06-30']

print("\nData from April to June:")
print(apr_to_jun.head())
print("...")
print(apr_to_jun.tail())
```

```python
jul_to_sep = df['2023-07-01':'2023-09-30']

print("\nData from July to September:")
print(jul_to_sep.head())
print("...")
print(jul_to_sep.tail())
```

```python
oct_to_dec = df['2023-10-01':'2023-12-31']

print("\nData from October to December:")
print(oct_to_dec.head())
print("...")
print(oct_to_dec.tail())
```

```python
def plot_interval(data, title):
    data['Value'].plot(title=title, figsize=(12, 6))
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.show()

# Plot for each interval
plot_interval(jan_to_mar, 'January to March')
plot_interval(apr_to_jun, 'April to June')
plot_interval(jul_to_sep, 'July to September')
plot_interval(oct_to_dec, 'October to December')
```

ðŸŒŸ Exercise 10 : Basic Visualization
Instructions
Use Matplotlib to visualize a simple dataset created with NumPy (e.g., a line graph of random numbers).

```python
x = np.arange(0, 100)
y = np.random.randn(100).cumsum()

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y, color='blue', linestyle='-', linewidth=2, label='Random Data')
ax.set_title('Line Graph of Random Numbers', fontsize=16)
ax.set_xlabel('X-axis (Index)', fontsize=12)
ax.set_ylabel('Y-axis (Cumulative Sum)', fontsize=12)
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
