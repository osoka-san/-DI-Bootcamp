```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']

temp_data = {
    'Tel Aviv': np.random.uniform(low=-5, high=35, size=12),
    'Miami': np.random.uniform(low=-5, high=35, size=12),
    'London': np.random.uniform(low=-5, high=35, size=12),
    'Paris': np.random.uniform(low=-5, high=35, size=12),
    'Tokyo': np.random.uniform(low=-5, high=35, size=12),
    'Berlin': np.random.uniform(low=-5, high=35, size=12),
    'New York': np.random.uniform(low=-5, high=35, size=12),
    'Moscow': np.random.uniform(low=-5, high=35, size=12),
    'Rome': np.random.uniform(low=-5, high=35, size=12),
    'Madrid': np.random.uniform(low=-5, high=35, size=12),
}

df = pd.DataFrame(temp_data, index=months)

print(df)
```

```python
annual_average_temp = df.mean(axis=0)
print(annual_average_temp)
```

```python
print(f"City with the highest annual temperature: {annual_average_temp.idxmax()}")
print(f"City with the lowest annual temperature: {annual_average_temp.idxmin()}")
```

```python
annual_average_temp.sort_values().plot(kind='bar', color='orange')

plt.title('Annual Average Temperature for Each City')
plt.xlabel('City')
plt.ylabel('Temperature (Â°C)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
