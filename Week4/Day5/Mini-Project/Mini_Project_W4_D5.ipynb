```python
import pandas as pd
df = pd.read_csv('/content/survey.csv')
df.info()
df.head()
```

Data Cleaning

```python
columns_to_drop = [
    'Timestamp', 'state', 'self_employed', 'comments', 'no_employees', 'remote_work',
    'tech_company', 'benefits', 'care_options', 'wellness_program', 'seek_help',
    'anonymity', 'leave', 'mental_health_consequence', 'phys_health_consequence',
    'coworkers', 'supervisor', 'mental_health_interview', 'phys_health_interview',
    'mental_vs_physical', 'obs_consequence'
]
df_cleaned = df.drop(columns=columns_to_drop)
df_cleaned['work_interfere'] = df_cleaned['work_interfere'].fillna('Unknown')
df_cleaned = df_cleaned[(df_cleaned['Age'] >= 18) & (df_cleaned['Age'] <= 100)]

def clean_gender(gender):
    gender = gender.lower()
    if 'female' in gender or 'f' in gender:
        return 'Female'
    elif 'male' in gender or 'm' in gender:
        return 'Male'
    else:
        return 'Other'

df_cleaned['Gender'] = df_cleaned['Gender'].apply(clean_gender)

print(df_cleaned.info())
print(df_cleaned.head())
```

What is the distribution of mental health conditions among different age groups in the tech industry?

```python
bins = [0, 25, 35, 45, 55, 65, 100]
labels = ['<25', '25-34', '35-44', '45-54', '55-64', '65+']
df_cleaned['age_group'] = pd.cut(df_cleaned['Age'], bins=bins, labels=labels, right=False)

age_group_distribution = df_cleaned.groupby('age_group')['treatment'].value_counts(normalize=True).unstack().fillna(0)

print(age_group_distribution)
```

How does the frequency of mental health issues vary by gender?

```python
gender_freq = df_cleaned.groupby('Gender')['treatment'].value_counts(normalize=True).unstack().fillna(0)

print(gender_freq)
```

Identify the countries with the highest and lowest reported rates of mental health issues in the tech industry.

```python
country_distribution = df_cleaned.groupby('Country')['treatment'].value_counts(normalize=True).unstack().fillna(0)
country_distribution['Rate'] = country_distribution['Yes']

highest_countries = country_distribution['Rate'].nlargest(5)
lowest_countries = country_distribution['Rate'].nsmallest(5)

print("Countries with highest rate:")
print(highest_countries)

print("\nCountries with lowest rate:")
print(lowest_countries)
```
