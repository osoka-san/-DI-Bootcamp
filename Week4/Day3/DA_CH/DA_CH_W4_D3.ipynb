```python
import matplotlib.pyplot as plt
from PIL import Image, ImageOps

def show_images(original, augmented, title):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(augmented)
    plt.title(title)
    plt.axis('off')

    plt.show()

image_path = '/content/00_010.png'
original_image = Image.open(image_path)

rotated_image = original_image.rotate(90)
show_images(original_image, rotated_image, "Rotated by 90 Degrees")

flipped_horizontally = ImageOps.mirror(original_image)
show_images(original_image, flipped_horizontally, "Flipped Horizontally")

flipped_vertically = ImageOps.flip(original_image)
show_images(original_image, flipped_vertically, "Flipped Vertically")

width, height = original_image.size
zoomed_image = original_image.resize((int(width * 1.2), int(height * 1.2)))
show_images(original_image, zoomed_image, "Zoomed In (1.2x)")
```
