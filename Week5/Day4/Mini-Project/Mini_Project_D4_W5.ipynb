```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = 'https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv'
penguins = pd.read_csv(url)

penguins.head()


```

```python

# Data Cleaning and Exploratory Analysis
# Checking for missing values in the dataset
missing_values = penguins.isnull().sum()

data_types = penguins.dtypes

# Checking for duplicates
duplicates = penguins.duplicated().sum()

missing_values, data_types, duplicates
```

```python
# Handling missing values
penguins_cleaned = penguins.dropna(subset=['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g'])

# Summary statistics after cleaning
summary_statistics = penguins_cleaned.describe(include='all')

# Display summary statistics
summary_statistics
```

```python
# Set the style for the plot
sns.set(style="whitegrid")

# Bar chart for species distribution
sns.countplot(data=penguins_cleaned, x='species', palette='Set2')
plt.title('Distribution of Penguin Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

The bar chart above shows the distribution of penguin species in the dataset. The majority of the penguins belong to the Adelie species, followed by Gentoo, and Chinstrap species.

```python
# Flipper Length vs. Body Mass: Scatter plot color-coded by species
sns.scatterplot(data=penguins_cleaned, x='flipper_length_mm', y='body_mass_g', hue='species', palette='Set2', s=100)
plt.title('Flipper Length vs Body Mass', fontsize=16)
plt.xlabel('Flipper Length (mm)', fontsize=12)
plt.ylabel('Body Mass (g)', fontsize=12)
plt.legend(title='Species')
plt.show()
```

Gentoo penguins tend to have larger body mass and longer flippers compared to the other species.
Adelie and Chinstrap penguins have similar body mass ranges but differ slightly in flipper length.


```python
# Island Distribution: Pie chart
penguins_cleaned['island'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
plt.title('Distribution of Penguins Across Islands', fontsize=16)
plt.ylabel('')  # Hide the y-label for aesthetic reasons
plt.show()
```

The pie chart above illustrates the distribution of penguins across the three islands. The majority of penguins are from Biscoe Island, followed by Dream and Torgersen.


```python
# Sex Distribution within Species: Grouped bar chart
sns.countplot(data=penguins_cleaned, x='species', hue='sex', palette='Set2')
plt.title('Sex Distribution within Each Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Sex')
plt.show()
```

The grouped bar chart above shows the distribution of sexes within each penguin species. Both male and female penguins are fairly evenly distributed across species, with the Gentoo species having a slightly higher count of males

```python
# Heatmap of Correlations between numerical variables
plt.figure(figsize=(10, 8))
# Include numeric_only=True to only include numerical columns in the correlation calculation
correlation_matrix = penguins_cleaned.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title('Correlation Heatmap of Numerical Variables', fontsize=16)
plt.show()
```

A strong positive correlation between flipper length and body mass.
Bill length and bill depth show a weak correlation, meaning they vary independently of each other.
Year has little to no correlation with the other measurements.

```python
# Boxplots for measurements (flipper length, body mass, bill measurements) segmented by species
plt.figure(figsize=(15, 10))
# Flipper Length
plt.subplot(2, 2, 1)
sns.boxplot(data=penguins_cleaned, x='species', y='flipper_length_mm', palette='Set2')
plt.title('Flipper Length by Species', fontsize=14)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Flipper Length (mm)', fontsize=12)

# Body Mass
plt.subplot(2, 2, 2)
sns.boxplot(data=penguins_cleaned, x='species', y='body_mass_g', palette='Set2')
plt.title('Body Mass by Species', fontsize=14)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Body Mass (g)', fontsize=12)

# Bill Length
plt.subplot(2, 2, 3)
sns.boxplot(data=penguins_cleaned, x='species', y='bill_length_mm', palette='Set2')
plt.title('Bill Length by Species', fontsize=14)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Bill Length (mm)', fontsize=12)

# Bill Depth
plt.subplot(2, 2, 4)
sns.boxplot(data=penguins_cleaned, x='species', y='bill_depth_mm', palette='Set2')
plt.title('Bill Depth by Species', fontsize=14)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Bill Depth (mm)', fontsize=12)

plt.tight_layout()
plt.show()
```

Flipper Length: Gentoo penguins have significantly longer flippers compared to Adelie and Chinstrap penguins.
Body Mass: Gentoo penguins are notably heavier, while Chinstrap penguins have the lowest body mass.
Bill Length: Gentoo penguins also have the longest bills, while Adelie penguins have the shortest.
Bill Depth: Adelie penguins have deeper bills compared to the other species.

```python
# Histogram for bill_length_mm
plt.subplot(1, 3, 1)
sns.histplot(penguins_cleaned['bill_length_mm'], kde=True, color='blue', bins=20)
plt.title('Distribution of Bill Length')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Frequency')

# Histogram for bill_depth_mm
plt.subplot(1, 3, 2)
sns.histplot(penguins_cleaned['bill_depth_mm'], kde=True, color='green', bins=20)
plt.title('Distribution of Bill Depth')
plt.xlabel('Bill Depth (mm)')
plt.ylabel('Frequency')

# Histogram for flipper_length_mm
plt.subplot(1, 3, 3)
sns.histplot(penguins_cleaned['flipper_length_mm'], kde=True, color='orange', bins=20)
plt.title('Distribution of Flipper Length')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

1.	Bill Length: Most penguins have a bill length between 35-50 mm, with a slight right skew, indicating some have longer bills.
2.	Bill Depth: Values range from 13-20 mm, with a more even distribution and a peak around 15-17 mm.
3.	Flipper Length: Most flipper lengths are concentrated between 190-210 mm, with the majority of penguins falling within this rang
